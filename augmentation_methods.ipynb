# Data Augmentation Example With MNIST Dataset Using Tensorflow

### Overview

Visualize data augmentation methods that relevant to MNIST datasets with using ImageDataGenerator.


```python
# Import mnist data
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# import matplotlib's pyplot library to visualize MNIST images
import matplotlib.pyplot as plt
%matplotlib inline

# import relevant library
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Helper function that visualizes the images befor being augmented

def visualise_raw(row_count, col_count):
    cells_count = row_count*col_count

    images = train_images[0:cells_count]
    labels = train_labels[0:cells_count]

    print("Raw Data")

    fig, axes = plt.subplots(row_count, col_count, figsize=(1.5*col_count,2*row_count))
    for i in range(cells_count):
        ax = axes[i//col_count, i%col_count]
        ax.imshow(images[i], cmap='gray_r')
        ax.set_title('Label: {}'.format(labels[i]))
    plt.tight_layout()
    plt.show()
```

```python
# Helper function that fits generated data and visualizes the result

def visualize_augmented(num_row, num_col, generated_data):
     num = num_row * num_col

     # fit the generator
     generated_data.fit(train_images.reshape(train_images.shape[0], 28, 28, 1))
     # plot after
     print('Augmented Data')
     fig2, axes2 = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
     for X, Y in generated_data.flow(train_images.reshape(train_images.shape[0], 28, 28, 1),train_labels.reshape(train_labels.shape[0], 1),batch_size=num,shuffle=False):
          for i in range(0, num):
               ax = axes2[i//num_col, i%num_col]
               ax.imshow(X[i].reshape(28,28), cmap='gray_r')
               ax.set_title('Label: {}'.format(int(Y[i])))
          break
     plt.tight_layout()
     plt.show()
```

```python
# Rotation

rotation_range_val = 90
rotated_data = ImageDataGenerator(rotation_range=rotation_range_val)

visualise_raw(2,8)
visualize_augmented(2,8,rotated_data)
```

```python
# Shift
# Translates the image vertically and horizontally
width_shift_val = 0.25
height_shift_val = 0.25

# transform
shifted_data = ImageDataGenerator(width_shift_range=width_shift_val, height_shift_range=height_shift_val)

visualise_raw(2,8)
visualize_augmented(2,8,shifted_data)

```

```python
# Shear
shear_range_val = 45

sheared_data = ImageDataGenerator(shear_range=shear_range_val)

visualise_raw(2,8)
visualize_augmented(2,8,sheared_data)
```

```python
# zoom
zoom_range_val = [0.5, 1.5]

zoomed_data = ImageDataGenerator(zoom_range=zoom_range_val)

visualise_raw(2,8)
visualize_augmented(2,8,zoomed_data)
```

```python
# Combine all transformations
rotation_range_val = 30
width_shift_val = 0.25
height_shift_val = 0.25
shear_range_val = 45
zoom_range_val = [0.5, 1.5]

augmented_data = ImageDataGenerator(rotation_range=rotation_range_val, width_shift_range=width_shift_val, height_shift_range=height_shift_val, shear_range=shear_range_val, zoom_range=zoom_range_val)

visualise_raw(4,8)
visualize_augmented(4,8,augmented_data)
```
